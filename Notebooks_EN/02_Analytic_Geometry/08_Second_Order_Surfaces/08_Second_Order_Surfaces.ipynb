### ***Task 3***

https://www.geogebra.org/calculator/kfdzbuhm

1. \( x^2 + y^2 + z^2 = 1 \),
2. \( (x - 1)^2 + y^2 + z^2 = 1 \).

* Expanding the second equation:

$$
(x - 1)^2 + y^2 + z^2 = 1
$$

* Expanding \( (x - 1)^2 \):

$$
(x^2 - 2x + 1) + y^2 + z^2 = 1
$$

* Simplifying:

$$
x^2 - 2x + 1 + y^2 + z^2 = 1
$$

* Substituting \( x^2 + y^2 + z^2 = 1 \):

$$
1 - 2x + 1 = 1
$$

* Simplifying further:

$$
2 - 2x = 1
$$

* Solving for \( x \):

$$
2x = 1 \quad \Rightarrow \quad x = \frac{1}{2}
$$

* Substituting \( x = \frac{1}{2} \) into \( x^2 + y^2 + z^2 = 1 \):

$$
\left( \frac{1}{2} \right)^2 + y^2 + z^2 = 1
$$

* Simplifying:

$$
\frac{1}{4} + y^2 + z^2 = 1
$$

* Thus:

$$
y^2 + z^2 = \frac{3}{4}
$$

* The equation of the intersection curve is:

$$
y^2 + z^2 = \frac{3}{4}
$$
