 2)Find the equation of a parabola intersecting the $Ox$ axis at points $x=2$, $x=4$, and passing through point $y(3)=1$.

 3)Find the center of the ellipse with the equation $x^2 + 4y^2 - 4x - 16y + 16 = 0$.

 4)Find the slope ($m>0$) of the line $y=mx-5$ that is tangent to the circle with the equation $x^2 + y^2=1$.

 5)Find the intersection points of the hyperbola $x^2 - y^2 = 1$ with the ellipse's line $x^2 + 4y^2 = 6$.

## ***Exercise 2***

Given $x=2$, $x=4$, and $y(3)=1$:

### 1. $y = (x-2)(x-4)$

### 2. Substituting $x=3$, $y(3)=1$:

$$
1 = a(3-2)(3-4)
$$

$$
1 = -1a \implies a = -1
$$

### 3. Substituting $a$:

$$
y = -1(x-2)(x-4)
$$

$$
y = -1(x^2 - 6x + 8) \implies {y = -x^2 + 6x - 8}

## ***Exercise 3***

Given $x^2 + 4y^2 - 4x - 16y + 16 = 0$:

$$
(x^2 - 4x) + 4(y^2 - 4y) = -16
$$

### 1. Completing the square:

$$
x^2 - 4x = (x-2)^2 - 4, \quad y^2 - 4y = (y-2)^2 - 4
$$

### 2. Substituting:

$$
(x-2)^2 - 4 + 4((y-2)^2 - 4) = -16
$$

$$
(x-2)^2 + 4(y-2)^2 - 20 = -16
$$

$$
(x-2)^2 + 4(y-2)^2 = 4
$$

### 3. Dividing through:

$$
\frac{(x-2)^2}{4} + \frac{(y-2)^2}{1} = 1
$$

The center of the ellipse is:

$$
{(2,2)}
$$

## ***Exercise 4***

The slope $(m > 0)$ of the line $y = mx - 5$ that is tangent to the circle $x^2 + y^2 = 1$:

### 1. Substituting $y = mx - 5$:

$$
x^2 + (mx - 5)^2 = 1
$$

$$
x^2 + m^2x^2 - 10mx + 25 = 1 \implies (1 + m^2)x^2 - 10mx + 24 = 0
$$

### 2. The discriminant $\Delta$:

$$
\Delta = (-10m)^2 - 4(1 + m^2)(24)
$$

$$
\Delta = 100m^2 - 96(1 + m^2) = 100m^2 - 96 - 96m^2
$$

$$
\Delta = 4m^2 - 96
$$

### 3. For tangency, $\Delta = 0$:

$$
4m^2 - 96 = 0 \implies m^2 = 24 \implies m = \sqrt{24} = {2\sqrt{6}}
$$

## ***Exercise 5***

For the hyperbola $x^2 - y^2 = 1$ and ellipse $x^2 + 4y^2 = 6$:

### 1. From $x^2 - y^2 = 1$:

$$
x^2 = y^2 + 1
$$

### 2. Substituting into $x^2 + 4y^2 = 6$:

$$
(y^2 + 1) + 4y^2 = 6
$$

$$
5y^2 + 1 = 6 \implies y^2 = 1 \implies y = \pm 1
$$

### 3. For $y = 1$:

$$
x^2 = 1^2 + 1 = 2 \implies x = \pm \sqrt{2}
$$

### 4. For $y = -1$:

$$
x^2 = (-1)^2 + 1 = 2 \implies x = \pm \sqrt{2}
$$

The intersection points are:

$$
{(\sqrt{2}, 1), (-\sqrt{2}, 1), (\sqrt{2}, -1), (-\sqrt{2}, -1)}
$$

## ***Exercise 6***

### For the hyperbola $x^2 - y^2 = 1$:

$$
\frac{x^2}{a^2} - \frac{y^2}{b^2} = 1, \quad a^2 = 1 \implies a = 1, \quad b^2 = 1 \implies b = 1
$$

### The hyperbola branches occur where $y = 0$:

$$
\frac{x^2}{1} = 1 \implies x = \pm 1
$$

### The distance between the branches is:

$$
\mathbf{Distance} = |1 - (-1)| = {2}
$$
