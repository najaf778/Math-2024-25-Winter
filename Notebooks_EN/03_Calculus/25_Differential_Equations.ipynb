## **25. Differential Equations**
1. Solve the following first-order ordinary differential equations:
   - $y'(x)= y$
   - $y'(x) = \frac{1}{2y(x)}$
  
2. Solve the first-order ordinary differential equations using the method of separation of variables for $x$ and $y=y(x)$:

   - $\frac{dy}{dx} = \frac{x}{y}$
   - $\frac{dy}{dx} = \frac{y}{x}$
   - $\frac{dy}{dx} = xy$

3. Solve the second-order ordinary differential equations:

   * $y''(x) + y'(x) = 0$, with boundary conditions $y(0) = 2$ and $y'(0) = -1$

   * $y''(x) - y(x)= 0$, with boundary conditions $y(0) = 2$ and $y'(0) = 0$

   * $\frac{d^2\,y(x)}{dx^2} = -\omega^2 y(x)$.

4. Check if the function $\psi(t, x) = A \cos(\omega t + kx)$ is a solution of the second-order partial differential equation (the so-called "wave equation"), where $v = \frac{\omega}{k}$:

$$
\frac{\partial^2 \psi(t, x)}{\partial t^2} - v^2 \frac{\partial^2 \psi(t, x)}{\partial x^2} = 0.
$$

## **Exercise 1**
### 1. $ y'(x) = y $

This differential equation can be solved by separation of variables.

**Step 1: Separate the variables**

$\frac{dy}{dx} = y \implies \frac{dy}{y} = dx$

**Step 2: Integrate both sides**

$\int \frac{1}{y} \, dy = \int 1 \, dx$

$\ln |y| = x + C_1$

**Step 3: Solve for $ y $**

$|y| = e^{x + C_1} = e^x e^{C_1}$

Let $ C_2 = e^{C_1} $, where $ C_2 $ is a positive constant:

$y = C_2 e^x$

Thus, the general solution is:

$y(x) = C_2 e^x$

where $C_2 $ is an arbitrary constant.

---

### 2. $ y'(x) = \frac{1}{2y(x)} $

This equation is also separable.

**Step 1: Separate the variables**

$\frac{dy}{dx} = \frac{1}{2y} \implies 2y \, dy = dx$

**Step 2: Integrate both sides**

$\int 2y \, dy = \int 1 \, dx$

$y^2 = x + C_3$

**Step 3: Solve for $ y $**

$y = \pm \sqrt{x + C_3}$

Hence, the general solution is:

$y(x) = \pm \sqrt{x + C_3}$

where $ C_3 $ is an arbitrary constant.

## **Exercise 2**
### 1. $ \frac{dy}{dx} = \frac{x}{y} $

$\frac{dy}{dx} = \frac{x}{y} \implies y \, dy = x \, dx$

$\int y \, dy = \int x \, dx$

$\frac{y^2}{2} = \frac{x^2}{2} + C_1$

$y^2 = x^2 + 2C_1$

Let $ C_2 = 2C_1 $, then:

$y^2 = x^2 + C_2$

$y = \pm \sqrt{x^2 + C_2}$

---

### 2. $ \frac{dy}{dx} = \frac{y}{x}$

$\frac{dy}{dx} = \frac{y}{x} \implies \frac{dy}{y} = \frac{dx}{x}$

$\int \frac{1}{y} \, dy = \int \frac{1}{x} \, dx$

$\ln |y| = \ln |x| + C_1$

$|y| = e^{\ln |x| + C_1} = e^{\ln |x|} e^{C_1} = |x| e^{C_1}$

Let $ C_2 = e^{C_1} $, then:

$y = C_2 x$

---

### 3. $ \frac{dy}{dx} = xy $

$\frac{dy}{dx} = xy \implies \frac{dy}{y} = x \, dx$

$\int \frac{1}{y} \, dy = \int x \, dx$

$\ln |y| = \frac{x^2}{2} + C_1$

$|y| = e^{\frac{x^2}{2} + C_1} = e^{\frac{x^2}{2}} e^{C_1}$

Let $ C_2 = e^{C_1} $, then:

$y = C_2 e^{\frac{x^2}{2}}$

## **Exercise 3**
### 1. $ y''(x) + y'(x) = 0 $ with boundary conditions $ y(0) = 2 $ and $ y'(0) = -1 $

$y''(x) + y'(x) = 0$

Let $ y'(x) = v(x) $, then $ v'(x) = y''(x) $:

$v'(x) + v(x) = 0$

This is a first-order linear ODE for $ v(x) $:

$\frac{dv}{dx} = -v \implies \frac{dv}{v} = -dx$

Integrating both sides:

$\ln |v| = -x + C_1 \implies v = C_1 e^{-x}$

Since $ v(x) = y'(x) $:

$y'(x) = C_1 e^{-x}$

Integrate again to find $ y(x) $:

$y(x) = -C_1 e^{-x} + C_2$

1. $ y(0) = 2 $:

   $-C_1 e^0 + C_2 = 2 \implies -C_1 + C_2 = 2 \implies C_2 = 2 + C_1$

2. $ y'(0) = -1 $:
   
   $C_1 e^0 = -1 \implies C_1 = -1$
   
   Using $ C_1 = -1 $ in $ C_2 = 2 + C_1 $:
   
   $C_2 = 2 + (-1) = 1$

$y(x) = e^{-x} + 1$

---

### 2. $ y''(x) - y(x) = 0 $ with boundary conditions $ y(0) = 2 $ and $ y'(0) = 0 $

$y''(x) - y(x) = 0$

The characteristic equation is:

$r^2 - 1 = 0 \implies r = \pm 1$

$y(x) = C_1 e^x + C_2 e^{-x}$

1. $ y(0) = 2 $:
   
   $C_1 e^0 + C_2 e^0 = 2 \implies C_1 + C_2 = 2$

2. $ y'(0) = 0 $:

   $C_1 e^0 - C_2 e^0 = 0 \implies C_1 - C_2 = 0 \implies C_1 = C_2$

   From $ C_1 + C_2 = 2 $ and $ C_1 = C_2 $:

   $2C_1 = 2 \implies C_1 = 1 \implies C_2 = 1$

$y(x) = e^x + e^{-x}$

---

### 3. $ \frac{d^2 y(x)}{dx^2} = -\omega^2 y(x) $

$y''(x) = -\omega^2 y(x)$

The characteristic equation is:

$r^2 + \omega^2 = 0 \implies r = \pm i\omega$

$y(x) = C_1 \cos(\omega x) + C_2 \sin(\omega x)$

## **Exercise 4**
Verify if the function $\psi(t, x) = A \cos(\omega t + kx)$ satisfies the wave equation:

$\frac{\partial^2 \psi(t, x)}{\partial t^2} - v^2 \frac{\partial^2 \psi(t, x)}{\partial x^2} = 0,$

where $v = \frac{\omega}{k}.$

1) Compute
$\psi(t, x) = A \cos(\omega t + kx)$

Differentiate with respect to $t$ once:

$\frac{\partial \psi(t, x)}{\partial t} = -A \omega \sin(\omega t + kx)$

Differentiate again with respect to $t$:

$\frac{\partial^2 \psi(t, x)}{\partial t^2} = -A \omega^2 \cos(\omega t + kx).$

2) Differentiate with respect to $x$ once:

$\frac{\partial \psi(t, x)}{\partial x} = -A k \sin(\omega t + kx)$

Differentiate again with respect to $x$:

$\frac{\partial^2 \psi(t, x)}{\partial x^2} = -A k^2 \cos(\omega t + kx)$

3) Substitute into the wave equation

$-A \omega^2 \cos(\omega t + kx) - v^2 (-A k^2 \cos(\omega t + kx)) = 0$

$-A \cos(\omega t + kx) \left( \omega^2 - v^2 k^2 \right) = 0$

4) Simplify the condition

Since $A \neq 0$ and $\cos(\omega t + kx) \neq 0$ (not identically zero),

$\omega^2 - v^2 k^2 = 0.$

$\omega^2 = v^2 k^2.$

From the definition of $v = \frac{\omega}{k}$,

we know that:

$v^2 = \frac{\omega^2}{k^2}$

Substitute:

$v^2 k^2 = \omega^2,$ the equation holds true.
