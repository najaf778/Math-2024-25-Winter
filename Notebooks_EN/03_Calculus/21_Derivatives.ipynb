## **Exercise 1**
### 1. $ y(x) = -3x + 3 $

The derivative of $ -3x+3 $ is calculated as follows:

$(-3x+3)' = (-3x)' + (3)' = (-3x)' + 0 = (-3)'x + (-3)x' = -3$.

Since the function is linear, the derivative of any term in the form $ ax + b $ is simply the coefficient $ a $.

$y'(x) = -3$

**Explanation**: The derivative of $ -3x $ is $ -3 $, while the derivative of the constant $ 3 $ is $ 0 $.

---
### 2. $ y(x) = \pi x + \sin(1) $

$y'(x) = (πx + \sin(1))' = (\pi x)' + (\sin(1))' = (\pi)'x + (\pi)x' + 0 = \pi$.

Here, $ \pi $ is a constant, as is $ \sin(1) $.

$y'(x) = \pi$

**Explanation**: The derivative of $ \pi x $ is $ \pi $, while the derivative of the constant $ \sin(1) $ is $ 0 $.

---
### 3. $ y(x) = 4 + \sin(2) $

$y(x)$ is a constant:

$4 + \sin(2) = c$

The derivative of a constant is $ 0 $:

$y'(x) = 0$

**Explanation**: Since the entire function is a constant, its derivative is zero.

---
### 4. $ y(x) = 2x^3 - 3x^2 + 8x - 9 $

We differentiate term by term using the **power rule**, which states that $ \frac{d}{dx} \left(x^n\right) = n \cdot x^{n-1} $:

$y'(x) = 6x^2 - 6x + 8$

---
### 5. $ y(x) = 6x^{1/3} $

Using the power rule:

$y'(x) = 6 \cdot \frac{1}{3} x^{-2/3} = 2x^{-2/3}$

**Explanation**: The derivative of $ x^{1/3} $ is $ \frac{1}{3}x^{-2/3} $, and multiplying by 6 gives $ 2x^{-2/3} $.

---
### 6. $ y(x) = \sqrt{x} $

Rewrite $ \sqrt{x} $ as $ x^{1/2} $:

$y'(x) = \frac{1}{2}x^{-1/2} = \frac{1}{2\sqrt{x}}$

**Explanation**: Using the power rule, the derivative of $ x^{1/2} $ is $ \frac{1}{2}x^{-1/2} $, which simplifies to $ \frac{1}{2\sqrt{x}} $.

---
### 7. $ y(x) = \cos(x) + \sin(x) $

Differentiate each trigonometric term:

$y'(x) = -\sin(x) + \cos(x)$

**Explanation**: The derivative of $ \cos(x) $ is $ -\sin(x) $, and the derivative of $ \sin(x) $ is $ \cos(x) $.

---
### 8. $ y(x) = 2\sin(x)\cos(x) $

This is a product of two functions, so we apply the **product rule**:

$y'(x) = 2\left( \cos(x) \cdot \cos(x) - \sin(x) \cdot \sin(x) \right) = 2\cos(2x)$

**Explanation**: Using the trigonometric identity $ 2\sin(x)\cos(x) = \sin(2x) $, we simplify the derivative to $ 2\cos(2x) $.

---
### 9. $ y(x) = x \sin(x) $

Apply the **product rule**:

$y'(x) = \sin(x) + x\cos(x)$

**Explanation**: The derivative of $ x $ is $ 1 $, giving $ \sin(x) $, while the derivative of $ \sin(x) $ is $ \cos(x) $, resulting in $ x\cos(x) $.

---
### 10. $ y(x) = (x+1)(x+1) = (x+1)^2 $

Using the power rule:

$y'(x) = 2(x+1)$

**Explanation**: The derivative of $ (x+1)^2 $ is $ 2(x+1) $, and the derivative of $ x+1 $ is $ 1 $.

---
### 11. $ y(x) = \frac{x}{x+1} $

Use the **quotient rule**, which states:

$$\frac{d}{dx}\left(\frac{u}{v}\right) = \frac{v u' - u v'}{v^2}$$

Substitute $ u = x $ and $ v = x+1 $:

$y'(x) = \frac{(x+1)(1) - x(1)}{(x+1)^2} = \frac{1}{(x+1)^2}$

**Explanation**: The derivatives of $ x $ and $ x+1 $ are both 1, resulting in the simplified fraction.

---
### 12. $ y(x) = (x+1)e^x $

Apply the product rule:

$y'(x) = e^x + (x+1)e^x = (x+2)e^x$

**Explanation**: The derivative of $ e^x $ is $ e^x $, and the derivative of $ x+1 $ is $ 1 $.

---
### 13. $ y(x) = \sin(x^2) $

Use the **chain rule**, which states:

$$\frac{d}{dx}\left(f(g(x))\right) = f'(g(x)) \cdot g'(x)$$

$y'(x) = 2x\cos(x^2)$

**Explanation**: The derivative of $ \sin(u) $ is $ \cos(u) $, and the derivative of $ x^2 $ is $ 2x $.

---
### 14. $ y(x) = e^{-2x} $

Using the chain rule:

$y'(x) = -2e^{-2x}$

**Explanation**: The derivative of $ e^u $ is $ e^u \cdot u' $, where $ u = -2x $, and $ u' = -2 $.

---
### 15. $ y(x) = \frac{1}{\sin(x+1)} $

Rewrite as $ y(x) = (\sin(x+1))^{-1} $ and apply the chain rule:

$y'(x) = -\frac{\cos(x+1)}{\sin^2(x+1)}$

**Explanation**: The derivative of $ \sin(x+1) $ is $ \cos(x+1) $, and we use the chain rule for the reciprocal.

---
### 16. $ y(x) = \sqrt{2x+1} $

Rewrite $ \sqrt{2x+1} $ as $ (2x+1)^{1/2} $, then use the chain rule:

$y'(x) = \frac{1}{2}(2x+1)^{-1/2} \cdot 2 = \frac{1}{\sqrt{2x+1}}$

**Explanation**: The derivative of $ (2x+1)^{1/2} $ is $ \frac{1}{2}(2x+1)^{-1/2} $, and multiplying by $ 2 $ simplifies the result.

$$\frac{d}{dx}\left(\ln(\sin(x))\right) = \cot(x)$$

**Reasoning**: The derivative of $ \ln(u) $ is $ \frac{1}{u}\cdot u' $. Let $ u = \sin(x) $:

$$\frac{d}{dx}\left(\ln(\sin(x))\right) = \frac{\cos(x)}{\sin(x)} = \cot(x)$$

## **Exercise 4**
1. $ \lim_{x \to 0} \frac{\sin(x)}{x} $

At $ x = 0 $, the expression is $ \frac{0}{0} $ (indeterminate). Apply **L'Hôpital's Rule**:

$\lim_{x \to 0} \frac{\sin(x)}{x} = \lim_{x \to 0} \frac{\cos(x)}{1} = 1$

2. $ \lim_{x \to \infty} \frac{\ln(x)}{x} $

For $ x \to \infty $, the form is $ \frac{\infty}{\infty} $. Apply L'Hôpital's Rule:

$\lim_{x \to \infty} \frac{\ln(x)}{x} = \lim_{x \to \infty} \frac{1/x}{1} = \lim_{x \to \infty} \frac{1}{x} = 0$

3. $ \lim_{x \to \infty} \frac{e^x}{x} $

As $ x \to \infty $, $ \frac{e^x}{x} \to \infty $, since $ e^x $ grows faster than $ x $.
