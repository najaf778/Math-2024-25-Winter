## **Exercise 6**
The profit function is given by $P(u) = -2u^2 + 50u - 300$.

The derivative is:

$P'(u) = -4u + 50$

Setting the derivative equal to zero to find the critical points:

$-4u + 50 = 0$

$-4u = -50$

$u = 12.5$

Thus, the number of units that maximizes the profit is $u = 12.5$.

## **Exercise 7**
### Largest Rectangle with 10 Meters of Perimeter
Let:
- $l$ represent the length of the rectangle
- $w$ represent the width of the rectangle

The perimeter is given by:

$P = 2l + 2w$

We know the total perimeter is 10 meters:

$2l + 2w = 10$

Simplifying:

$l + w = 5$

The area $A$ of the rectangle is:

$A = l \cdot w$

Express $l$ in terms of $w$ using $l + w = 5$:

$l = 5 - w$

Substitute into the area formula:

$A(w) = (5 - w)w = 5w - w^2$

To maximize the area, differentiate $A(w)$:

$A'(w) = \frac{d}{dw}(5w - w^2) = 5 - 2w$

Set the derivative equal to zero:

$5 - 2w = 0$

$2w = 5$

$w = 2.5$

From $l + w = 5$, we find:

$l = 5 - 2.5 = 2.5$

Thus, the dimensions of the rectangle that maximize the area are:

Length = 2.5 meters, Width = 2.5 meters.

## **Exercise 8**
Consider the function $f(x) = x^2 + 3x - 5$.

The derivative is:

$f'(x) = 2x + 3$

Setting $f'(x) = 0$ to find critical points:

$2x + 3 = 0$

$x = -\frac{3}{2}$

Now, compute the second derivative:

$f''(x) = 2$

Since $f''(x) > 0$, the function is concave up at $x = -\frac{3}{2}$, indicating a local minimum.

Evaluate the function at $x = -\frac{3}{2}$:

$f\left(-\frac{3}{2}\right) = \left(-\frac{3}{2}\right)^2 + 3\left(-\frac{3}{2}\right) - 5$

$= \frac{9}{4} - \frac{9}{2} - 5$

$= \frac{9}{4} - \frac{18}{4} - \frac{20}{4}$

$= -\frac{29}{4}$

Thus, the critical point is $x = -\frac{3}{2}$, and the minimum value of the function is $f(-\frac{3}{2}) = -\frac{29}{4}$.

## **Exercise 9**
Finding Extremum of the Function

$f(x) = \frac{x^2 + 2x + 1}{x - 1}$

Differentiate using the quotient rule:

$f'(x) = \frac{(2x + 2)(x - 1) - (x^2 + 2x + 1)(1)}{(x - 1)^2}$

Simplify the numerator:

$(2x + 2)(x - 1) = 2x(x - 1) + 2(x - 1) = 2x^2 - 2x + 2x - 2 = 2x^2 - 2$

Numerator becomes:

$(2x^2 - 2) - (x^2 + 2x + 1) = 2x^2 - 2 - x^2 - 2x - 1 = x^2 - 2x - 3$

Thus:

$f'(x) = \frac{x^2 - 2x - 3}{(x - 1)^2}$

To find critical points, set $f'(x) = 0$:

$\frac{x^2 - 2x - 3}{(x - 1)^2} = 0$

$x^2 - 2x - 3 = 0$

Factorize:

$(x - 3)(x + 1) = 0$

$x = 3 \quad \text{or} \quad x = -1$

The domain of $f(x)$ excludes $x = 1$ because the denominator becomes zero.

Thus:

- Critical points: $x = 3$ and $x = -1$
- Vertical asymptote at $x = 1$
