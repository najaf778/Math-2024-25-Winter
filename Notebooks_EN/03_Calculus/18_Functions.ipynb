## **Exercise 3**
Let $f(x) = e^x$ and $g(x) = \ln(x)$. Check $f(g(x))$ and $g(f(x))$.

1. Substituting $g(x) = \ln(x)$ into $f(x)$:  

$
f(g(x)) = f(\ln(x)) = e^{\ln(x)}
$

$
e^{\ln(x)} = x \quad \text{for } x > 0
$

We conclude:  

$
f(g(x)) = x \quad \text{for } x > 0
$

2. Substituting $f(x) = e^x$ into $g(x)$:  

$
g(f(x)) = g(e^x) = \ln(e^x)
$


$
\ln(e^x) = x \quad \text{for all real } x
$

concluding:  

$
g(f(x)) = x
$

**Conclusion:**  
- $f(g(x)) = x$ for $x > 0$
- $g(f(x)) = x$ for all real $x$

**We Notice:**  

The functions $f(x) = e^x$ and $g(x) = \ln(x)$ are reverse operations of one another.
