## **Exercise 1**
- $ f(x) = \cos(x) $ around $ x = 0 $ up to the 4th degree

### Step 1: Compute the Derivatives of $ f(x) = \cos(x) $

1. $ f(x) = \cos(x) $

   $ f(0) = \cos(0) = 1 $

2. **First derivative**:

   $ f'(x) = -\sin(x) $

   $ f'(0) = -\sin(0) = 0 $

3. **Second derivative**:

   $ f''(x) = -\cos(x) $

   $ f''(0) = -\cos(0) = -1 $

4. **Third derivative**:

   $ f'''(x) = \sin(x) $

   $ f'''(0) = \sin(0) = 0 $

5. **Fourth derivative**:

   $ f^{(4)}(x) = \cos(x) $

   $ f^{(4)}(0) = \cos(0) = 1 $

### Step 2:

The Taylor series for $ f(x) $ around $ x = 0 $ is:

$$
f(x) = f(0) + \frac{f'(0)}{1!}x + \frac{f''(0)}{2!}x^2 + \frac{f'''(0)}{3!}x^3 + \frac{f^{(4)}(0)}{4!}x^4 + \cdots
$$

Substituting the derivatives:

$$
\cos(x) \approx 1 - \frac{x^2}{2} + \frac{x^4}{24}
$$

- $ h(x) = \frac{1}{1-x} $ around $ x = 0 $ up to the 4th degree.

### Step 1

1. $ h'(x) = \frac{1}{(1-x)^2} \quad \Rightarrow \quad h'(0) = 1 $

2. $ h''(x) = \frac{2}{(1-x)^3} \quad \Rightarrow \quad h''(0) = 2 $

3. $ h'''(x) = \frac{6}{(1-x)^4} \quad \Rightarrow \quad h'''(0) = 6 $

4. $ h^{(4)}(x) = \frac{24}{(1-x)^5} \quad \Rightarrow \quad h^{(4)}(0) = 24 $

### Taylor Series Formula

$$
h(x) = 1 + \frac{h'(0)}{1!}x + \frac{h''(0)}{2!}x^2 + \frac{h'''(0)}{3!}x^3 + \frac{h^{(4)}(0)}{4!}x^4
$$

Substituting the derivatives:

$$
h(x) \approx 1 + x + x^2 + x^3 + x^4
$$

- $ g(x) = \sin(x) $ around $ x = \pi $ up to the 4th degree.

### Step 1: Computing the Derivatives of $ g(x) = \sin(x) $

1. $ g'(x) = \cos(x) $

2. $ g''(x) = -\sin(x) $

3. $ g'''(x) = -\cos(x) $

4. $ g^{(4)}(x) = \sin(x) $

### Step 2: Evaluating the Derivatives at $ x = \pi $

1. $ g(\pi) = \sin(\pi) = 0 $

2. $ g'(\pi) = \cos(\pi) = -1 $

3. $ g''(\pi) = -\sin(\pi) = 0 $

4. $ g'''(\pi) = -\cos(\pi) = 1 $

5. $ g^{(4)}(\pi) = \sin(\pi) = 0 $

### Step 3: Writing the Taylor Series Formula

The Taylor series around $ x = \pi $ is:

$$
g(x) = g(\pi) + \frac{g'(\pi)}{1!}(x - \pi) + \frac{g''(\pi)}{2!}(x - \pi)^2 + \frac{g'''(\pi)}{3!}(x - \pi)^3 + \frac{g^{(4)}(\pi)}{4!}(x - \pi)^4
$$

Substituting the values:

$$
g(x) \approx -(x - \pi) + \frac{1}{6}(x - \pi)^3
$$

## **Exercise 2**

### Tangent Line to $ f(x) = e^{\sin(x)} $ at $ x_0 = \pi $

### Step 1: Compute $ f(x_0) $

$$
f(x) = e^{\sin(x)}
$$

$$
f(\pi) = e^{\sin(\pi)} = e^0 = 1
$$

### Step 2: Compute $ f'(x) $

$$
f'(x) = e^{\sin(x)} \cdot \cos(x)
$$

$$
f'(\pi) = e^{\sin(\pi)} \cdot \cos(\pi) = e^0 \cdot (-1) = -1
$$

### Step 3: Writing the Equation of the Tangent Line

The tangent line formula is:

$$
y = f'(x_0)(x - x_0) + f(x_0)
$$

Substituting $ x_0 = \pi $, $ f'(x_0) = -1 $, and $ f(x_0) = 1 $:

$$
y = -1 \cdot (x - \pi) + 1
$$

$$
y = -x + \pi + 1
$$
